```python
import scrapy
from scrapy.http import HtmlResponse
from scrapy import Selector

from scrapy.crawler import CrawlerProcess

import requests
```

```python
# dataclass

class MovieItem(scrapy.Item):
    title = scrapy.Field()
    rating = scrapy.Field()
    summary = scrapy.Field()
    genre = scrapy.Field()
    runtime = scrapy.Field()
    directors = scrapy.Field()
    writers = scrapy.Field()
    cast = scrapy.Field()

class ActorItem(scrapy.Item):
    actor_name = scrapy.Field()
    character = scrapy.Field()
```

```python
class ImdbSpider(scrapy.Spider):
    name = "imdb"
    allowed_domains = ["imdb.com"]
    base_url = "https://imdb.com"
    start_urls = ['https://www.imdb.com/chart/top',]
   
    def parse(self, response):
        table_rows = response.xpath(
            './/*[@class="chart full-width" and @data-caller-name="chart-top250movie"]/'
            'tbody[@class="lister-list"]/tr'
        
        )
        for row in table_rows:
            rating = row.xpath("./td[@class='ratingColumn imdbRating']/strong/text()").extract_first(),
            rel_url = row.xpath("td[@class='titleColumn']/a/@href").extract_first().strip()
            row_url = self.base_url + rel_url
            yield scrapy.Request(row_url, callback=self.parseOneMovie, meta={'rating' : rating})
    
    def parseOneMovie(self, response):
        item = MovieItem()
        item["rating"] = response.meta["rating"][0]
        item['title'] = response.xpath('.//*/div[@class="title_wrapper"]/h1/text()').extract_first().strip()
        item["summary"] = response.xpath(".//*/div[@class='summary_text']/text()").extract_first().strip()
        item['directors'] = response.xpath('.//*/div[@class="credit_summary_item"]/a/text()').extract_first().strip()
        item['writers'] = response.xpath('.//*/div[@class="credit_summary_item"]/a/text()').extract()[1].strip()

        item["cast"] = list()

        for cast in response.xpath(".//table[@class='cast_list']/tr")[1:]:
            actor = ActorItem()
            
            actor["actor_name"] = cast.xpath("./td[2]/a/text()").extract_first().strip()
            actor["character"] = cast.xpath("./td[@class='character']/a/text()").extract_first()
            
            item["cast"].append(actor)

        return item
```

```python
%%time
process = CrawlerProcess()

process.crawl(ImdbSpider)
process.start()
```

```python
import scrapy
from scrapy.http import HtmlResponse
from scrapy import Selector

from scrapy.crawler import CrawlerProcess

import requests
```

```python
# dataclass

class MovieItem(scrapy.Item):
    title = scrapy.Field()
#     rating = scrapy.Field()
#     summary = scrapy.Field()
#     genre = scrapy.Field()
#     runtime = scrapy.Field()
#     directors = scrapy.Field()
#     writers = scrapy.Field()
    cast = scrapy.Field()
    url = scrapy.Field()

class ActorItem(scrapy.Item):
#     actor_name = scrapy.Field()
#     character = scrapy.Field()
    i = scrapy.Field()
    bio = scrapy.Field()
    born = scrapy.Field()
    movies = scrapy.Field()
    name = scrapy.Field()
    url = scrapy.Field()
    
class ImdbSpider(scrapy.Spider):
    name = "imdb"
    allowed_domains = ["imdb.com"]
    base_url = "https://imdb.com"
#     start_urls = ['https://www.imdb.com/chart/top',]
    start_urls = ['https://www.imdb.com/search/name/?gender=male%2Cfemale&ref_=nv_cel_m',]
    base_url_www = "https://www.imdb.com"
    
    def parse(self, response):
        table_rows = response.xpath('.//*[@id="content-2-wide" and @class="redesign"]').\
                        xpath('.//*[@class="lister-list"]/div')
        
        i = 0
        for row in table_rows:
            i += 1
#             rating = row.xpath("./td[@class='ratingColumn imdbRating']/strong/text()").extract_first(),
            rel_url = row.xpath('.//*[@class="lister-item-content"]/h3/a/@href').extract_first().strip() + '/'
            row_url = self.base_url + rel_url
#             yield scrapy.Request(row_url, callback=self.parseOneMovie, meta={'rating' : rating})
            yield scrapy.Request(row_url, callback=self.parseOneActor,\
#                                  meta={'url' : row_url, 'dont_redirect': True})
                                 meta={'url' : row_url, 'i': i})
    
    def parseOneActor(self, response):
        item = ActorItem()
#         item['i'] =  response.meta['i']
        item['url'] = response.meta['url']
        
        item['bio'] = response.xpath('.//*[@id="name-bio-text"]').\
            xpath('.//*[@class="inline"]/text()').extract_first().strip()
        
        try:
            item['born'] = response.xpath('//*[@id="name-born-info"]/time/@datetime').extract_first().strip()
        except:
            item['born'] = None
        
        item['name'] = response.xpath('.//*[@id="name-overview-widget-layout"]').\
                            xpath('.//h1/span[1]/text()').extract_first().strip()

        item['movies'] = list()
        
        for movie_i in response.xpath('.//*[@id="filmography"]').\
                        xpath('//*[@class="filmo-category-section"][1]/div')[:15]:
            
            movie_name = movie_i.xpath("b/a/text()").extract_first().strip()
            item['movies'].append(movie_name)
            
            rel_movie_url = movie_i.xpath("b/a/text()").extract_first().strip()
            item['movies'].append(movie_name)
            
            
            movie_rel_url = movie_i.xpath("b/a/@href").extract_first().strip()
            movie_url = self.base_url_www + movie_rel_url
            
            yield scrapy.Request(movie_url, callback=self.parseMovieInfo,\
#                                  meta={'url' : row_url, 'dont_redirect': True})
                                 meta={'url' : movie_url,\
                                       'title': movie_name})
            
#         return item
        
    def parseMovieInfo(self, response):

        item = MovieItem()

        item['title'] = response.meta['title']
        item['url'] = response.meta['url']

        item['cast'] = list()

        actor_table = response.xpath('.//*[@class="cast_list"]/tr')

        for actor_i in actor_table[1:-1]:
            actor_i_name = actor_i.xpath('td[2]/a/text()').extract_first().strip()
            item['cast'].append(actor_i_name)

        return item
        
#     def parseOneMovie(self, response):
#         item = MovieItem()
#         item["rating"] = response.meta["rating"][0]
#         item['title'] = response.xpath('.//*/div[@class="title_wrapper"]/h1/text()').extract_first().strip()
#         item["summary"] = response.xpath(".//*/div[@class='summary_text']/text()").extract_first().strip()
#         item['directors'] = response.xpath('.//*/div[@class="credit_summary_item"]/a/text()').extract_first().strip()
#         item['writers'] = response.xpath('.//*/div[@class="credit_summary_item"]/a/text()').extract()[1].strip()

#         item["cast"] = list()

#         for cast in response.xpath(".//table[@class='cast_list']/tr")[1:]:
#             actor = ActorItem()
            
#             actor["actor_name"] = cast.xpath("./td[2]/a/text()").extract_first().strip()
#             actor["character"] = cast.xpath("./td[@class='character']/a/text()").extract_first()
            
#             item["cast"].append(actor)

#         return item
```
